## Implementação dos Algoritmos

### Geração de números aleatórios

```python
import random

constant_100     = 10**2
constant_1000    = 10**3
constant_10000   = 10**4
constant_100000  = 10**5
constant_1000000 = 10**6

def generate_random_numbers(quantity):
    random_numbers = []
    for _ in range(quantity):
        random_numbers.append(random.randint(1, constant_1000000))
    return random_numbers
```

## Merge Sort

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    return merge(left_half, right_half)

def merge(left, right):
    merged = []
    left_index = right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    merged.extend(left[left_index:])
    merged.extend(right[right_index:])

    return merged
```

### Execução

```python
values = generate_random_numbers(constant_100)
sorted_values = merge_sort(values)

print("Desordenado:", values)
print("Ordenado:", sorted_values)
```

## Quick Sort

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
```

### Execução

```python
values = generate_random_numbers(constant_100)
sorted_values = quick_sort(values)

print("Desordenado:", values)
print("Ordenado:", sorted_values)
```

## Bubble Sort

```python
def bubble_sort(arr):
    sorted_arr = arr[:]
    n = len(sorted_arr)

    for i in range(n):
        for j in range(0, n - i - 1):
            if sorted_arr[j] > sorted_arr[j + 1]:
                sorted_arr[j], sorted_arr[j + 1] = sorted_arr[j + 1], sorted_arr[j]

    return sorted_arr
```

### Execução

```python
values = generate_random_numbers(constant_100)
sorted_values = bubble_sort(values)

print("Desordenado:", values)
print("Ordenado:", sorted_values)
```

### Melhor Caso

#### Bubble Sort

```python
def generate_sorted_numbers(quantity):
    return list(range(1, quantity + 1))

generated_sorted_numbers = generate_sorted_numbers(constant_10000)
generated_random_numbers = generate_random_numbers(constant_10000)
```

```python
def bubble_sort_best(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

inicio_bubble_melhor = time.perf_counter()
bubble_sort_best(generated_sorted_numbers)
fim_bubble_melhor = time.perf_counter()

print(f"Tempo de execução do Bubble Sort (melhor caso): {fim_bubble_melhor - inicio_bubble_melhor:.6f} segundos")
```

#### Quick Sort

```python
def quick_sort_worst(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) - 1] 
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort_worst(left) + middle + quick_sort_worst(right)

inicio_quick_pior = time.perf_counter()
quick_sort_worst(generated_random_numbers)
fim_quick_pior = time.perf_counter()

print(f"Tempo de execução do Quick Sort (pior caso): {fim_quick_pior - inicio_quick_pior:.6f} segundos")
```

## Consumo de Memória

```python
import tracemalloc

algoritmos = {
    "Quick Sort": (quick_sort, "O(log n)"),
    "Merge Sort": (merge_sort, "O(n)"),
    "Bubble Sort": (bubble_sort, "O(1)")
}

resultados = []

for nome, (funcao, espaco) in algoritmos.items():
    tracemalloc.start()

    funcao(values)

    memoria_atual, memoria_pico = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    resultados.append([
        nome,
        espaco,
        round(memoria_pico / 1024, 2)
    ])

pd.DataFrame(
    resultados,
    columns=[
        "Algoritmo",
        "Complexidade de Espaço",
        "Pico de Memória (KB)"
    ]
)
```

## Análise Amortizada

Neste trabalho, a análise amortizada não se aplica aos algoritmos estudados,pois Bubble Sort, Merge Sort e Quick Sort
são algoritmos de ordenação e não estruturas de dados dinâmicas.
Por esse motivo, a comparação foi feita considerando a complexidade assintótica, o consumo de memória e o tempo de execução de cada algoritmo.

## Testes e Medição de Desempenho

```python
import time

algoritmos = {
    "Quick Sort": quick_sort,
    "Merge Sort": merge_sort,
    "Bubble Sort": bubble_sort
}

tamanhos = [constant_100, constant_1000, constant_10000, constant_100000, constant_1000000]

resultados = []

for tamanho in tamanhos:
    values = generate_random_numbers(tamanho)

    for nome, funcao in algoritmos.items():
        if nome == "Bubble Sort" and tamanho > constant_10000:
            continue

        inicio = time.perf_counter()

        funcao(values)

        fim = time.perf_counter()

        resultados.append([
            nome,
            tamanho,
            round((fim - inicio), 4)
        ])

df = pd.DataFrame(
    resultados,
    columns=[
        "Algoritmo",
        "Quantidade de Elementos",
        "Tempo (ms)"
    ]
)

tabela = df.pivot(
    index="Algoritmo",
    columns="Quantidade de Elementos",
    values="Tempo (ms)"
)

tabela.columns = [f"{col} elementos" for col in tabela.columns]

tabela = tabela.round(4)

tabela
```

```python
import matplotlib.pyplot as plt

tabela.T.plot(marker="o", figsize=(8, 5))

plt.title("Comparação de Desempenho dos Algoritmos")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Tempo de Execução (ms)")
plt.grid(True)

plt.show()
```

## Conclusão

Com base nas análises realizadas, conclui-se que o Bubble Sort apresentou o pior desempenho para entradas maiores, devido à sua complexidade O(n²). Já o Merge Sort e o Quick Sort obtiveram resultados significativamente melhores, mantendo um crescimento de tempo compatível com O(n log n). Os testes práticos confirmaram a teoria estudada, demonstrando a importância da escolha adequada do algoritmo de ordenação conforme o tamanho da entrada e os recursos disponíveis.
